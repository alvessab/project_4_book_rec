```python
import pandas as pd
import numpy as np

from langdetect import detect
```

```python
df = pd.read_csv("../data/goodreads_final.csv")
df.head()
```

```python
df["lang"] = df.bookDesc.apply(lambda x: detect(x))
df.head()
```

```python
df.lang.value_counts()
```

```python
#save df with language column as csv
df.to_csv("../data/goodreads_final_w_lang.csv", index=False)
```

```python
#drop non-english columns
df_sub = df.loc[df.lang == "en"]
df_sub.bookTitle.nunique()
```

```python
df_sub.info()
```

```python
#see duplicates
duplicateRows = df_sub[df_sub.duplicated(['bookTitle'])]
duplicateRows
```

```python
df_sub.loc[df_sub.bookTitle == "Apollyon"]
```

```python
df_sub2 = df_sub.drop_duplicates(subset=['bookTitle'])
df_sub2.info()
```

```python
#save df with language column as csv
df_sub2.to_csv("../data/goodreads_final_cleaned.csv", index=False)
```

```python
books = ['Twilight', 'The Hunger Games', 'To Kill a Mockingbird']
```

```python
df_sub3 = df_sub2.loc[df_sub2.bookTitle.isin(books)]
```

```python
df_sub4= df_sub2.loc[~df_sub2.bookTitle.isin(books)]
df_sub5 = df_sub4.sample(n=5000-len(books),  random_state=42).reset_index(drop=True)
df_sub5= df_sub5.append(df_sub3)
df_sub5.info()
```

```python
#save df with language column as csv
df_sub5.to_csv("../data/goodreads_final_seven.csv", index=False)
```
